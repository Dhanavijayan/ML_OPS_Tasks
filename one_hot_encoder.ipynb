```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

df = pd.read_csv("supermarket_sample.csv")
```

```python
data = df.copy()

categorical_cols = ["Device_Type"] if "Device_Type" in df.columns else []
categorical_cols = ["Theft_Level"]

dummies = pd.get_dummies(data[categorical_cols], dtype=int)
merged = pd.concat([data.drop(columns=categorical_cols), dummies], axis="columns")

X = merged.drop(columns=["Row_ID", "Shrinkage_INR"])
y = merged["Shrinkage_INR"]

model = LinearRegression()
model.fit(X, y)

print(model.score(X, y))
```

```python
dfle = df.copy()

le = LabelEncoder()
for col in dfle.select_dtypes(include="object").columns:
    if col != "Shrinkage_INR":
        dfle[col] = le.fit_transform(dfle[col])

y = dfle["Shrinkage_INR"].values

X_raw = dfle.drop(columns=["Row_ID", "Shrinkage_INR"]).values

ct = ColumnTransformer(
    [("theft", OneHotEncoder(drop="first", sparse_output=False), [0])],
    remainder="passthrough"
)

X = ct.fit_transform(X_raw)

model.fit(X, y)

X_pred = ct.transform(X_raw[:1])
print(model.predict(X_pred))
```
